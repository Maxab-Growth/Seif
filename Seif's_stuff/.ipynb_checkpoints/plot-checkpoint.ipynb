```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from common_functions import ret_metabase
```

```python
df = ret_metabase("egypt", 62578)
```

```python


# Sample DataFrame structure
# You should replace this with your actual DB query result
# For example: df = pd.read_sql(query, connection)
# Optional: Sort by total combined points
df = df.sort_values(by='TOTAL_COMBINED_LOYALTY_POINTS', ascending=False)

# Plot setup
plt.figure(figsize=(14, 7))
bar_width = 0.3
x = range(len(df))

# Bar plots for each points category
plt.bar(x, df['TOTAL_ECOM_LOYALTY_POINTS'], width=bar_width, label='eCom Points', color='skyblue')
plt.bar([p + bar_width for p in x], df['FINTECH_LOYALTY_POINTS'], width=bar_width, label='Fintech Points', color='salmon')
plt.bar([p + 2 * bar_width for p in x], df['TOTAL_COMBINED_LOYALTY_POINTS'], width=bar_width, label='Total Points', color='seagreen')

# Labeling
plt.xlabel('Retailer ID')
plt.ylabel('Loyalty Points')
plt.title('Loyalty Points Breakdown per Retailer')
plt.xticks([p + bar_width for p in x], df['RETAILER_ID'])
plt.legend()
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
