```python
from common_functions import google_sheets, ret_metabase
```

```python
def assign_data_by_mapping(df, mapping_df):
    # store retail-agent mapping in a dictionary 
    mapping_dict = dict(zip(mapping_df['main_system_id'], mapping_df['agent_assigned']))
    
    
    assigned_agents = []
    # missing_retailers = [] could use it later for viewing errors

    for retailer_id in df['retailer_id']:
        if retailer_id in mapping_dict:
            assigned_agents.append(mapping_dict[retailer_id])
        else:
            assigned_agents.append(None)
    
    df['agent_assigned'] = assigned_agents

    df = df.reset_index(drop=True)

    return df
```

```python
def get_mapping():
    mapping_data = google_sheets("MA Telesales Cycle Assignment", "Sheet1", "get")
    agent_group = google_sheets("Fintech Runsheet", "Agent_to_group", "get")

    mapping_data.drop(columns=["retailer_id"], inplace=True)

    agent_group = agent_group[['Agent_name', 'Group_type']]

    agent_group = agent_group[agent_group['Agent_name'] != 'Benchmark']

    # Normalize column name
    agent_group = agent_group.rename(columns={'Group_type': 'group_type'})

    # Merge and rename Agent_name to agent_assigned
    merged_df = mapping_data.merge(agent_group, on='group_type', how='left')
    merged_df = merged_df.rename(columns={'Agent_name': 'agent_assigned'})
    merged_df.drop(columns=["group_type"], inplace=True)
    return merged_df
```

```python
Retention_dfs = ret_metabase('MOROCCO', 18879)
Retention_dfs.columns = map(str.lower, Retention_dfs.columns)
```

```python
mapped_agents = get_mapping()
```

```python
final_data = assign_data_by_mapping(Retention_dfs, mapped_agents)
```
