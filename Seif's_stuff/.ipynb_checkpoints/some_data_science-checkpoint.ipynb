```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Feature Columns
FEATURE_COLS = [
    'TOTAL_THP', 'TOTAL_THP_BP', 'POS_THP', 'APP_THP',
    'TXNS_LAST_30_DAYS', 'DAYS_ACTIVE_LAST_30_DAYS',
    'TENURE_MONTHS', 'DAYS_SINCE_LAST_TXN', 'ECOM_TAG'
]

# 1. Prepare and scale data
df_clean = final_sample.copy()  # Should be 12,000 rows
df_clean[FEATURE_COLS] = df_clean[FEATURE_COLS].fillna(0)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_clean[FEATURE_COLS])

# 2. KMeans Clustering
kmeans = KMeans(n_clusters=8, random_state=42, n_init='auto')
df_clean['cluster'] = kmeans.fit_predict(X_scaled)

# 3. Force exactly 1500 rows per cluster
balanced_dfs = []
max_per_cluster = 1500

for label in sorted(df_clean['cluster'].unique()):
    cluster_data = df_clean[df_clean['cluster'] == label]
    if len(cluster_data) >= max_per_cluster:
        sampled = cluster_data.sample(n=max_per_cluster, random_state=42)
    else:
        # Oversample with replacement if not enough
        sampled = cluster_data.sample(n=max_per_cluster, replace=True, random_state=42)
    balanced_dfs.append(sampled)

df_balanced = pd.concat(balanced_dfs).reset_index(drop=True)

# 4. Assign agent tags
agent_tag_map = {
    0: 'C1',
    1: 'T1',
    2: 'T2',
    3: 'T3',
    4: 'T4',
    5: 'T5',
    6: 'T6',
    7: 'T7'
}
df_balanced['agent_tag'] = df_balanced['cluster'].map(agent_tag_map)

# 5. PCA Visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(scaler.transform(df_balanced[FEATURE_COLS]))

plt.figure(figsize=(10, 7))
scatter = plt.scatter(
    X_pca[:, 0],
    X_pca[:, 1],
    c=df_balanced['agent_tag'].astype('category').cat.codes,
    cmap='tab10'
)
plt.title("Retailer Clusters by Agent Assignment (Balanced KMeans w/ Oversampling)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.grid(True)
plt.legend(*scatter.legend_elements(), title="Agent Tags", loc='best', labels=sorted(agent_tag_map.values()))
plt.tight_layout()
plt.show()

# 6. Export + Validate
df_balanced[['RETAILER_ID', 'agent_tag']].to_csv("retailers_balanced_kmeans.csv", index=False)
print(df_balanced['agent_tag'].value_counts())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# -----------------------------
# 1. Configuration
# -----------------------------
FEATURE_COLS = [
    'TOTAL_THP', 'TOTAL_THP_BP', 'POS_THP', 'APP_THP',
    'TXNS_LAST_30_DAYS', 'DAYS_ACTIVE_LAST_30_DAYS',
    'TENURE_MONTHS', 'DAYS_SINCE_LAST_TXN', 'ECOM_TAG'
]

AGENT_TAG_MAP = {
    0: 'C1',
    1: 'T1',
    2: 'T2',
    3: 'T3',
    4: 'T4',
    5: 'T5',
    6: 'T6',
    7: 'T7'
}

N_CLUSTERS = 8
TARGET_PER_CLUSTER = 1500

# -----------------------------
# 2. Load and Prepare Data
# -----------------------------
df_clean = final_sample.copy()
df_clean[FEATURE_COLS] = df_clean[FEATURE_COLS].fillna(0)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_clean[FEATURE_COLS])

# -----------------------------
# 3. Apply KMeans Clustering
# -----------------------------
kmeans = KMeans(n_clusters=N_CLUSTERS, random_state=42, n_init='auto')
df_clean['cluster'] = kmeans.fit_predict(X_scaled)

# -----------------------------
# 4. Balance Clusters via Oversampling
# -----------------------------
balanced_dfs = []

for cluster_label in sorted(df_clean['cluster'].unique()):
    cluster_df = df_clean[df_clean['cluster'] == cluster_label]
    if len(cluster_df) >= TARGET_PER_CLUSTER:
        sampled_df = cluster_df.sample(n=TARGET_PER_CLUSTER, random_state=42)
    else:
        sampled_df = cluster_df.sample(n=TARGET_PER_CLUSTER, replace=True, random_state=42)
    balanced_dfs.append(sampled_df)

df_balanced = pd.concat(balanced_dfs).reset_index(drop=True)

# -----------------------------
# 5. Assign Agent Tags
# -----------------------------
df_balanced['agent_tag'] = df_balanced['cluster'].map(AGENT_TAG_MAP)

# -----------------------------
# 6. Visualize with PCA
# -----------------------------
pca = PCA(n_components=2)
X_pca = pca.fit_transform(scaler.transform(df_balanced[FEATURE_COLS]))

plt.figure(figsize=(10, 7))
scatter = plt.scatter(
    X_pca[:, 0], X_pca[:, 1],
    c=df_balanced['agent_tag'].astype('category').cat.codes,
    cmap='tab10'
)
plt.title("Retailer Clusters by Agent Assignment (Balanced)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.legend(*scatter.legend_elements(), title="Agent Tags", loc='best', labels=sorted(AGENT_TAG_MAP.values()))
plt.grid(True)
plt.tight_layout()
plt.show()

# -----------------------------
# 7. Export Results
# -----------------------------
df_balanced[['RETAILER_ID', 'agent_tag']].to_csv("retailers_balanced_kmeans.csv", index=False)

# -----------------------------
# 8. Check Final Distribution
# -----------------------------
print(df_balanced['agent_tag'].value_counts())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# --- Config ---
FEATURE_COLS = [
    'TOTAL_THP', 'TOTAL_THP_BP', 'POS_THP', 'APP_THP',
    'TXNS_LAST_30_DAYS', 'DAYS_ACTIVE_LAST_30_DAYS',
    'TENURE_MONTHS', 'DAYS_SINCE_LAST_TXN', 'ECOM_TAG'
]
TAG_LIST = ['T1', 'T2', 'T3', 'T4', 'T5', 'T6', 'T7', 'C1']
MAX_PER_TAG = 1500

# --- Load and prep ---
df = final_sample.copy()
df[FEATURE_COLS] = df[FEATURE_COLS].fillna(0)
df['agent_tag'] = 'Unassigned'

# --- Rule-based tagging (strict limit per tag) ---
rule_conditions = {
    'T1': (df['POS_THP'] > 2000) & (df['TXNS_LAST_30_DAYS'] > 10),
    'T2': (df['APP_THP'] > 2500) & (df['TXNS_LAST_30_DAYS'] > 10),
    'T3': (df['TOTAL_THP_BP'] > 2000) & (df['TXNS_LAST_30_DAYS'] > 5),
    'T4': (df['TOTAL_THP'] > 1000) & (df['ECOM_TAG'] == 1),
    'T5': (df['DAYS_SINCE_LAST_TXN'] > 45),
    'T6': (df['TENURE_MONTHS'] > 12) & (df['TXNS_LAST_30_DAYS'] < 3),
    'T7': df['TXNS_LAST_30_DAYS'].between(3, 5),
    'C1': (df['TENURE_MONTHS'] <= 2) & (df['TXNS_LAST_30_DAYS'] <= 2)
}

for tag, condition in rule_conditions.items():
    unassigned_mask = df['agent_tag'] == 'Unassigned'
    eligible = df[unassigned_mask & condition].copy()
    assign_n = min(MAX_PER_TAG, len(eligible))
    selected = eligible.sample(n=assign_n, random_state=42)
    df.loc[selected.index, 'agent_tag'] = tag

# --- KMeans to fill under-assigned tags ---
current_counts = df['agent_tag'].value_counts()
needed = {tag: MAX_PER_TAG - current_counts.get(tag, 0) for tag in TAG_LIST}
needed = {tag: n for tag, n in needed.items() if n > 0}

remaining = df[df['agent_tag'] == 'Unassigned'].copy()
if not remaining.empty and needed:
    # Fit scaler and KMeans
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(remaining[FEATURE_COLS])
    
    kmeans = KMeans(n_clusters=len(needed), random_state=42)
    remaining['cluster'] = kmeans.fit_predict(X_scaled)
    
    cluster_to_tag = dict(zip(sorted(remaining['cluster'].unique()), list(needed.keys())))
    
    to_append = []
    for clust, tag in cluster_to_tag.items():
        group = remaining[remaining['cluster'] == clust]
        take_n = min(len(group), needed[tag])
        sampled = group.sample(n=take_n, random_state=42)
        sampled['agent_tag'] = tag
        to_append.append(sampled)
    
    df.loc[pd.concat(to_append).index, 'agent_tag'] = pd.concat(to_append)['agent_tag']

# --- Fill any remaining gaps with random sampling ---
current_counts = df['agent_tag'].value_counts()
needed = {tag: MAX_PER_TAG - current_counts.get(tag, 0) for tag in TAG_LIST}
needed = {tag: n for tag, n in needed.items() if n > 0}

remaining = df[df['agent_tag'] == 'Unassigned'].copy()
for tag, n_needed in needed.items():
    if len(remaining) == 0 or n_needed == 0:
        break
    take_n = min(n_needed, len(remaining))
    sampled = remaining.sample(n=take_n, random_state=42)
    df.loc[sampled.index, 'agent_tag'] = tag
    remaining = df[df['agent_tag'] == 'Unassigned'].copy()

# --- Final Output ---
final = df[df['agent_tag'] != 'Unassigned'].copy().reset_index(drop=True)

# Validate
print(final['agent_tag'].value_counts())

# Optional save
final[['RETAILER_ID', 'agent_tag']].to_csv("retailers_balanced_rule_kmeans.csv", index=False)
```
