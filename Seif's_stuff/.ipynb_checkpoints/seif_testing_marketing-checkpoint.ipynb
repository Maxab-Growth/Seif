```python
from common_functions import google_sheets



import pytz
from datetime import datetime
import pandas as pd
import numpy as np

marketing_data = google_sheets('Marketing test', 'Sheet1', 'get')
Agents = google_sheets('Marketing test', 'Sheet2', 'get')

marketing_data = marketing_data[marketing_data['is_retailer'].str.lower() == 'yes'].copy()
# Force mobile_number to string and remove scientific notation
marketing_data['mobile_number'] = marketing_data['mobile_number'].astype(str).str.replace('.0', '', regex=False)
selected_columns = ['mobile_number', 'full_name', 'المحافظة', 'created_time']
marketing_data = marketing_data[selected_columns]

marketing_data['description'] = (
    'Project: POS Acquisition + Marketing Lead + العميل عمل طلب ماكينة يوم: ' +
    pd.to_datetime(marketing_data['created_time']).dt.strftime('%Y-%m-%d')
)

marketing_data = marketing_data[['mobile_number', 'full_name', 'المحافظة', 'description']]



av_agents = Agents[Agents['Available'] == 'yes'].copy()
av_agents = av_agents[["Agent_assigned"]]
# print(marketing_data["mobile_number"])

pos_sheet = google_sheets("POS Test", "Sheet1", "get")
pos_sheet['Mobile Number'] = pos_sheet['Mobile Number'].astype(str).str.replace('.0', '', regex=False)
# print(pos_sheet["Mobile Number"])
numbers_called  = pos_sheet["Mobile Number"] 
new_clients = marketing_data[~marketing_data['mobile_number'].isin(numbers_called)]
print("there is :", len(new_clients))

num_retailers = len(new_clients)
num_agents = len(av_agents)

agent_names = av_agents['Agent_assigned'].values

# Repeat the agent list enough times, then trim to match length exactly
repeated_agents = np.tile(agent_names, int(np.ceil(num_retailers / len(agent_names))))[:num_retailers]

# Shuffle for randomness
np.random.shuffle(repeated_agents)

# Assign to DataFrame
new_clients['Name'] = repeated_agents

new_clients = new_clients.rename(columns={
    'mobile_number': 'Mobile Number',
    'description': 'Description'
})

new_clients = new_clients[['Name','Mobile Number', 'full_name', 'المحافظة', 'Description']]
print(new_clients)

google_sheets("POS Test", "Sheet1", "append", df=new_clients )
```
