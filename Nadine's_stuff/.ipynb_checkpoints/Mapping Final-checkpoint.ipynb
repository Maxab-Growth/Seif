```python
# Refined function to parse both WKB and GML formats
def parse_geometry(row):
    try:
        if row['WKB'] == 1:  # If it's WKB
            try:
                # Attempt to load WKB data
                return wkb.loads(row['COORDINATES'], hex=True)
            except Exception as e:
                print(f"Error parsing WKB in row {row.name}: {e}")
                return None
        else:  # If it's GML
            try:
                # Extract and parse coordinates from GML
                coords_match = re.search(r'<coordinates>(.*?)</coordinates>', row['COORDINATES'])
                if coords_match:
                    coords_text = coords_match.group(1)
                    # Convert coordinate pairs into floats
                    coords = [tuple(map(float, coord.split(','))) for coord in coords_text.strip().split()]
                    return Polygon(coords)
                else:
                    print(f"Invalid GML format in row {row.name}")
                    return None
            except Exception as e:
                print(f"Error parsing GML in row {row.name}: {e}")
                return None
    except Exception as e:
        print(f"Unexpected error in row {row.name}: {e}")
        return None

# Apply the refined parsing function to your DataFrame
df['geometry'] = df.apply(parse_geometry, axis=1)

# Convert to a GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry='geometry', crs="EPSG:4326")

# Print the number of rows in gdf to confirm that it was successfully created
print(f"Number of valid geometries: {len(gdf)}")

gdf.head()
```

```python
# Define bins and ordered labels for NMV Market Share
bins = [0, 0.25, 0.5, 0.75, 1.0]
labels = ['0 to 25%', '25% to 50%', '50% to 75%', '75% to 100%']
gdf['MarketShareCategory'] = pd.cut(gdf['NMV_MARKET_SHARE'], bins=bins, labels=labels, right=False, ordered=True)

gdf['NMV_MARKET_SHARE_PERCENT'] = gdf['NMV_MARKET_SHARE'] * 100


# Update the color mapping to reflect the desired legend order
color_discrete_map = {
    '0 to 25%': 'red',
    '25% to 50%': 'yellow',
    '50% to 75%': 'lightgreen',
    '75% to 100%': 'green'
}

# Specify category order for the legend
category_order = [
    '0 to 25%', 
    '25% to 50%', 
    '50% to 75%', 
    '75% to 100%'
]

# Create the Plotly choropleth map with ordered discrete colors
fig = px.choropleth_mapbox(
    gdf,
    geojson=gdf.geometry.__geo_interface__,
    locations=gdf.index,
    color='MarketShareCategory',
    color_discrete_map=color_discrete_map,
    # animation_frame='MONTH',
    mapbox_style="carto-positron",
    center={"lat": 31.5, "lon": -7.1},
    hover_data={"DISTRICT_NAME": True, "NMV_MARKET_SHARE_PERCENT": ":.2f", "TOTAL_NMV": True, "BRAND_NMV": True,  "BRAND_ACTIVE_USERS": True},
    zoom=6,
    title="NMV Market Share by District",
    category_orders={"MarketShareCategory": category_order}  # Enforce the legend order
)
```

```python
# Apply the refined parsing function to your DataFrame
districts['geometry'] = districts.apply(parse_geometry, axis=1)

# Convert to a GeoDataFrame
gdf2 = gpd.GeoDataFrame(districts, geometry='geometry', crs="EPSG:4326")

# Print the number of rows in gdf to confirm that it was successfully created
print(f"Number of valid geometries: {len(gdf2)}")

gdf2.head()
```
